```python
results = {
    "100": 23.4,
    "110": 24.3,
    "120": 25,
    "130": 25.7,
    "140": 26.6,
    "150": 26.9,
    "160": 27.0,
    "170": 27.0,
}
```

```python
import matplotlib.pyplot as plt

# Extract x and y values
x = [int(power) for power in results.keys()]
y = list(results.values())

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'bo-')  # Blue dots connected by lines

# Add labels and title
plt.xlabel('Power Limit (w)')
plt.ylabel('TFLOPS')
plt.title('RTX 3060 Power Limit vs FLOPs')

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
# Save the plot
plt.savefig('3060.png')
```
